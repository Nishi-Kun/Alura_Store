### Importação dos dados



```python
#importando dados em CSV com a biblioteca pandas:

import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#0. Importando bibliotecas de relatórios visuais:

```python
import matplotlib.pyplot as plt
```

#1. Análise do faturamento


```python
#Criando função para calcular o faturamento da loja:
calcular_faturamento = lambda lista: round(sum(lista),2)

#Utilizando a função para calcular o faturamento de cada loja:
faturamento_loja1 = calcular_faturamento(loja['Preço'])
faturamento_loja2 = calcular_faturamento(loja2['Preço'])
faturamento_loja3 = calcular_faturamento(loja3['Preço'])
faturamento_loja4 = calcular_faturamento(loja4['Preço'])

#Separando categorias e valores:
faturamentos = [faturamento_loja1,faturamento_loja2,faturamento_loja3,faturamento_loja4]
faturamento_lojas = {f'Loja {i+1}':faturamento for i, faturamento in enumerate(faturamentos) }
lojas = faturamento_lojas.keys()
```

```python
#Gerando gráfico de colunas para analisar os faturamentos por loja:
plt.bar(lojas,faturamentos, color='mediumseagreen')

#Aplicando rótulos de dados:
for i,fatu in enumerate(faturamentos):
  plt.text(i,fatu + 0.5, str(f'R${fatu}'), ha='center', va='bottom')

# Título e títulos dos eixos:
plt.title('Faturamento Por Lojas')
plt.xlabel('Lojas')
plt.ylabel('Faturamento')

# Exibir gráfico:
plt.show()
```

# 2. Vendas por Categoria


```python
#Separando categorias:
categoria_produtos = list(set(loja['Categoria do Produto']))
categorias_vendidas_loja1 = list(loja['Categoria do Produto'])
categorias_vendidas_loja2 = list(loja2['Categoria do Produto'])
categorias_vendidas_loja3 = list(loja3['Categoria do Produto'])
categorias_vendidas_loja4 = list(loja4['Categoria do Produto'])

#Quantidade de vendas por categoria, em cada loja:
vendas_loja1 = [categorias_vendidas_loja1.count(categoria) for categoria in categoria_produtos]
vendas_loja2 = [categorias_vendidas_loja2.count(categoria) for categoria in categoria_produtos]
vendas_loja3 = [categorias_vendidas_loja3.count(categoria) for categoria in categoria_produtos]
vendas_loja4 = [categorias_vendidas_loja4.count(categoria) for categoria in categoria_produtos]
```

```python
def gerar_gráfico_barras(categoria,valor,nloja):
  #Gerando gráfico de barras para analisar as vendas por categoria de cada loja:
  plt.barh(categoria,valor, color = 'Mediumseagreen')

  #Aplicando rótulos de dados:
  for i, valor in enumerate(valor):
      plt.text(valor + 0.5, i, str(valor), va='center')

  # Título e títulos dos eixos:
  plt.title(f'Vendas Por Categoria - Loja {nloja}')
  plt.xlabel('Quantidade De Vendas')
  plt.ylabel('Categorias')

  # Exibir gráfico:
  plt.show()

gerar_gráfico_barras(categoria_produtos,vendas_loja1,1)
gerar_gráfico_barras(categoria_produtos,vendas_loja2,2)
gerar_gráfico_barras(categoria_produtos,vendas_loja3,3)
gerar_gráfico_barras(categoria_produtos,vendas_loja4,4)
```

# 3. Média de Avaliação das Lojas

```python
#Separando notas de cada loja:
avaliacoes_loja1 = list(loja['Avaliação da compra'])
avaliacoes_loja2 = list(loja2['Avaliação da compra'])
avaliacoes_loja3 = list(loja3['Avaliação da compra'])
avaliacoes_loja4 = list(loja4['Avaliação da compra'])

#Função para calcular media de uma lista:
def calcular_media(lista)-> int:
  media = sum(lista)/len(lista)
  return round(media,2)

#Calculando media com a função:
avaliacao_media_loja1 = calcular_media(avaliacoes_loja1)
avaliacao_media_loja2 = calcular_media(avaliacoes_loja2)
avaliacao_media_loja3 = calcular_media(avaliacoes_loja3)
avaliacao_media_loja4 = calcular_media(avaliacoes_loja4)

#lista de medias:
medias = [avaliacao_media_loja1,avaliacao_media_loja2,avaliacao_media_loja3,avaliacao_media_loja4]
print(list(zip(lojas,medias)))
```

```python
# Função para contar quantas avaliações de cada nota (1 a 5)
def contar_avaliacoes(lista):
    contagem = [lista.count(nota) for nota in range(1, 6)]
    return contagem

# Função para gerar gráficos de pizza das lojas:
def gerar_grafico_pizza(avaliacoes, nome_loja, posicao):
    contagem = contar_avaliacoes(avaliacoes)
    labels = [f'Nota {i}' for i in range(1, 6)]
    cores = ['red', 'orange', 'yellow', 'lightgreen', 'green']

    plt.subplot(2, 2, posicao)  # 2 linhas, 2 colunas, posição atual
    plt.pie(contagem, labels=labels, autopct='%1.1f%%', startangle=90, colors=cores)
    plt.title(f'{nome_loja}')

# Plotando os 4 gráficos
plt.figure(figsize=(7, 7))

gerar_grafico_pizza(avaliacoes_loja1, 'Loja 1', 1)
gerar_grafico_pizza(avaliacoes_loja2, 'Loja 2', 2)
gerar_grafico_pizza(avaliacoes_loja3, 'Loja 3', 3)
gerar_grafico_pizza(avaliacoes_loja4, 'Loja 4', 4)

plt.tight_layout()
plt.show()
```

# 4. Produtos Mais e Menos Vendidos

```python
# Separando produtos vendidos de cada loja:
produtos_loja1 = list(loja['Produto'])
produtos_loja2 = list(loja2['Produto'])
produtos_loja3 = list(loja3['Produto'])
produtos_loja4 = list(loja4['Produto'])
produtos = list(set(produtos_loja1))

#função para contar quantidade de produtos:
def contagem_itens(lista1,lista2):
  for item_unico in produtos:
    contagem = [lista2.count(item) for item in lista1]
    return contagem
#criando lista de tuplas que contém quantidade e nome do produto:
produtos_vendidos_loja1 = list(zip(produtos,contagem_itens(produtos,produtos_loja1)))
produtos_vendidos_loja2 = list(zip(produtos,contagem_itens(produtos,produtos_loja2)))
produtos_vendidos_loja3 = list(zip(produtos,contagem_itens(produtos,produtos_loja3)))
produtos_vendidos_loja4 = list(zip(produtos,contagem_itens(produtos,produtos_loja4)))

#Função para verificar produto mais vendido de cada loja:
def verificar_mais_vendido(lista):
  mais_vendido = max(lista, key=lambda x: x[1])
  return mais_vendido

#Função para verificar produto menos vendido de cada loja:
def verificar_menos_vendido(lista):
  menos_vendido = min(lista, key=lambda x: x[1])
  return menos_vendido

#criando dicionário para a análise de cada loja e produto:
analise_loja1 = {'Produtos Mais Vendidos':verificar_mais_vendido(produtos_vendidos_loja1),'Produtos Menos Vendidos': verificar_menos_vendido(produtos_vendidos_loja1) }
analise_loja2 = {'Produtos Mais Vendidos':verificar_mais_vendido(produtos_vendidos_loja2),'Produtos Menos Vendidos': verificar_menos_vendido(produtos_vendidos_loja2) }
analise_loja3 = {'Produtos Mais Vendidos':verificar_mais_vendido(produtos_vendidos_loja3),'Produtos Menos Vendidos': verificar_menos_vendido(produtos_vendidos_loja3) }
analise_loja4 = {'Produtos Mais Vendidos':verificar_mais_vendido(produtos_vendidos_loja4),'Produtos Menos Vendidos': verificar_menos_vendido(produtos_vendidos_loja4) }

#imprimindo análise realizada:
print(f'Análise loja1: {analise_loja1}\nAnálise loja2: {analise_loja2}\nAnálise loja3: {analise_loja3}\nAnálise loja4: {analise_loja4}')


```

# 5. Frete Médio por Loja


```python
#calculando frete por loja:
frete_medio_loja1 = calcular_media(list(loja['Frete']))
frete_medio_loja2 = calcular_media(list(loja2['Frete']))
frete_medio_loja3 = calcular_media(list(loja3['Frete']))
frete_medio_loja4 = calcular_media(list(loja4['Frete']))

analise_frete = {'Frete_Médio_loja1':frete_medio_loja1,'Frete_Médio_loja2':frete_medio_loja2,'Frete_Médio_loja3':frete_medio_loja3,'Frete_Médio_loja4':frete_medio_loja4,}

print(analise_frete)

print(f'\nA loja que possui o menor valor de frete médio é a LOJA 4')
```
